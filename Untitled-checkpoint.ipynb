```python
plt.title("Box Plot of Average Price")
ax = sns.boxplot(dfa["AveragePrice"])
```

```python
plt.figure(figsize=(16,8))
plt.title("Type vs Average Price")
ax = sns.boxplot(y = "type", x = "AveragePrice", data = dfa, palette = 'pink')
```

```python
conventional_avo = dfa[dfa["type"].isin(['conventional'])]
conventional_avo = conventional_avo.sort_values(by='AveragePrice')
ax = sns.factorplot('AveragePrice','region', data=conventional_avo,
                    hue='year',
                    height=13,
                    aspect=0.8,
                    palatte = 'magma',
                    join=False
                   )
```
